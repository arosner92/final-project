```python
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.neural_network import MLPRegressor
from sklearn.externals import joblib
import numpy as np
import time
import pickle
import pandas as pd
import sklearn
from sklearn import linear_model

data = pd.read_csv('Resources/House_prices_train_FINAL.csv')

data.head()
```

```python
y = data['SalePrice']
x = data[['LotArea','TotalSF','WasRemodeled','IsNew','BedroomAbvGr','TotalBathrooms','OverallCond','ExterCond','HasBasement','HasGarage','HasPorch','HasPool','KitchenQual','FireplaceQu','GarageFinish','BsmtFinType2']]
x.head()
```

```python
x = x.apply(pd.to_numeric, errors='coerce')
y = y.apply(pd.to_numeric, errors='coerce')
x.fillna(0, inplace=True)
y.fillna(0, inplace=True)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.5, random_state = 42)
```

```python
reglinear = linear_model.LinearRegression()
reglinear.fit(X_train,y_train)
```

```python
print(reglinear.coef_)
```

```python
y_pred = reglinear.predict(X_test)
print(y_pred)
```

```python
print(y_test)
```

```python
print(np.mean((y_test-y_pred)**2))
```

```python
print(y_test - y_pred)
```

```python

pickle.dump(reglinear, open('model.pkl','wb'))


# Loading model to compare the results

model = pickle.load(open('model.pkl','rb'))

print(model.predict([[1200,850, 0,1,3,2.5,5,3,1,1,1,0,4,0,2,1]]))
```
